Дана страница https://ru.wikipedia.org/wiki/Премия_«Оскар»_за_лучший_фильм#Достижения_по_другим_номинациям_Оскара

Заметьте, на страничке эта информация в нескольких таблицах. По данным ответьте на вопросы. P.S. Результат нужно именно посчитать, если на странице в явном виде указан ответ на вопрос, просто выгрузить этот ответ нельзя:

    1. Сколько раз были номинированы фильмы, созданные при участии Стивена Спилберга?
    2. Фильмы какой компании получили больше Оскаров за лучший фильм: Warner Bros. или Paramount Pictures и насколько больше?

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup


link = 'https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%B5%D0%BC%D0%B8%D1%8F_%C2%AB%D0%9E%D1%81%D0%BA%D0%B0%D1%80%C2%BB_%D0%B7%D0%B0_%D0%BB%D1%83%D1%87%D1%88%D0%B8%D0%B9_%D1%84%D0%B8%D0%BB%D1%8C%D0%BC#%D0%94%D0%BE%D1%81%D1%82%D0%B8%D0%B6%D0%B5%D0%BD%D0%B8%D1%8F_%D0%BF%D0%BE_%D0%B4%D1%80%D1%83%D0%B3%D0%B8%D0%BC_%D0%BD%D0%BE%D0%BC%D0%B8%D0%BD%D0%B0%D1%86%D0%B8%D1%8F%D0%BC_%D0%9E%D1%81%D0%BA%D0%B0%D1%80%D0%B0'
response = re.sub(r', ', '; ', requests.get(link).text)
soup = BeautifulSoup(response, 'lxml')
tables = soup.find_all('table')
# print(len(tables))
# print(tables[10])
```

```python
f = open('table_3_1.csv', 'w', encoding = 'utf8')
for i in range(10):
    rows = tables[i].find_all('tr')
    
    for row in rows[1:]:
    #     row = re.sub(r', ', '; ', row)
        data = row.find_all('td')
        i = 0
        if int(data[0].attrs.get('rowspan', 0)) != 0:
            tmp = data[0].text.strip()
            i = 1
        f.write(f'{tmp},{data[i].text.strip()},{data[i + 1].text.strip()},\
                {data[i + 2].text.strip()}, \n')

f.close()

df = pd.read_csv('table_3_1.csv', encoding='utf8', header=None)
df.rename(columns = {0:'Year', 1:'Film', 2:'Company', 3:'Producer'}, inplace = True)
df = df.fillna('')
df
```

На данный момент есть готовый dataframe со всей информацией.

Осталось извлечь из него информацию.

```python
df_spilberg = df[df['Producer'].str.contains("Стивен Спилберг")]

df_paramount = df[df['Company'].str.contains("Paramount Pictures")]
df_warner = df[df['Company'].str.contains("Warner Bros")]

ln_paramount = len(df_paramount)
ln_warner = len(df_warner)

# print(f'Фильмы Warner Bros. - {ln_warner}\nФильмы Paramount Pictures - {ln_paramount}') 

print('1. Сколько раз были номинированы фильмы, созданные при участии \
Стивена Спилберга?\nОтвет:', len(df_spilberg)) 
print('2. Фильмы какой компании получили больше Оскаров за лучший фильм: \
Warner Bros. или Paramount Pictures и насколько больше?\nОтвет: Warner Bros. на', ln_warner - ln_paramount)
```

## Задача 2

Используя удобные для вас модули Python (Например requests, bs4 и pandas, но можно и любые другие) сгрузить в удобную для вас структуру данных (например pandas-табличку) данные о номинантах и победителях премии Оскар за лучшую женскую роль за всю историю существования премии. Заметьте, на страничке эта информация в нескольких таблицах. По данным ответьте на вопросы. P.S. Результат нужно именно посчитать, если на странице в явном виде указан ответ на вопрос, просто выгрузить этот ответ нельзя:

    1. Какая актриса чаще всего номинировалась на Оскар?
    2. Насколько отличается количество номинаций самой часто номинировавшейся актрисы и второй по частоте?

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup


link = 'https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%B5%D0%BC%D0%B8%D1%8F_%C2%AB%D0%9E%D1%81%D0%BA%D0%B0%D1%80%C2%BB_%D0%B7%D0%B0_%D0%BB%D1%83%D1%87%D1%88%D1%83%D1%8E_%D0%B6%D0%B5%D0%BD%D1%81%D0%BA%D1%83%D1%8E_%D1%80%D0%BE%D0%BB%D1%8C'
response = requests.get(link).text
soup = BeautifulSoup(response, 'lxml')
tables = soup.find_all('table')
# print(len(tables))
# print(tables[11])
```

```python
f = open('table_3_2.csv', 'w', encoding = 'utf8')
for i in range(1, 12):
    rows = tables[i].find_all('tr')
    for row in rows[1:]:
        data = row.find_all('td')
        if len(data) == 5:
            tmp0 = data[0].text.strip()
            tmp1 = data[2].text.strip()
            i = 3
        elif len(data) == 3:
            tmp1 = data[0].text.strip()
            i = 1
        elif len(data) == 2:
            i = 0 

        f.write(f'{tmp0}`{tmp1}`{data[i].text.strip()}`\
                {data[i + 1].text.strip()}` \n')

f.close()

df = pd.read_csv('table_3_2.csv', encoding='utf8', header=None, sep='`')
df.rename(columns = {0:'Ceremony', 1:'Actress', 2:'Film', 3:'Role'}, inplace = True)
df
```

На данный момент есть готовый dataframe со всей информацией.

Осталось извлечь из него информацию.

```python
res = df['Actress'].value_counts()
#print(res,'\n\n')

print('1. Какая актриса чаще всего номинировалась на Оскар?\nОтвет:', res.index[0][2:])
print('2. Насколько отличается количество номинаций самой часто\
номинировавшейся актрисы и второй по частоте?\nОтвет:', res[0] - res[1])
```

## Задача 3

Дана страница https://www.worldometers.info/world-population/world-population-by-year/

На ней представлены данные о росте численности населения Земли. Используя удобные для вас модули Python (Например requests, bs4 и pandas, но можно и любые другие) сгрузить в удобную для вас структуру данных (например pandas-табличку) необходимые данные и нарисовать графики ежегодного прироста городского населения и ежегодного прироста общего населения, начиная с 1951 года.

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup


link = 'https://www.worldometers.info/world-population/world-population-by-year/'
response = requests.get(link).text
soup = BeautifulSoup(response, 'lxml')
tables = soup.find_all('table')
# print(len(tables))
# print(tables[0])
```

```python
f = open('table_3_3.csv', 'w', encoding = 'utf8')
rows = tables[0].find_all('tr')

for row in rows[1:]:
    data = row.find_all('td')
    for i in range(0, 7):
        st = re.sub(r'[,%]', r'', data[i].text.strip())
        f.write(f'{st};')
    f.write('\n')
f.close()

df = pd.read_csv('table_3_3.csv', encoding='utf8', header=None, sep=';')

df.rename(columns = {0:'Year', 1:'World Population', 2:'Yearly Change %'\
, 3:'Net Change', 4:'Density (P/Km²)', 5:'Urban Pop', 6:'Urban Pop % ', 7:'Urban Change'}, inplace = True)
df['Urban Change'] = df['Urban Pop'].diff(periods=-1)
df = df[:70]
df
```

На данный момент есть готовый dataframe со всей информацией.

Осталось построить графики.

```python
p = df.plot(x='Year', y=['Net Change', 'Urban Change'])
```
